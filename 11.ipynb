```python
#データのハッシュ値を計算するための標準ライブラリ
#代表的な関数：MD5, SHA-1, SHA-256, SHA-512 など

#hashlib → そのバイト列の MD5 を計算する

import hashlib


#-----------------チェックサム-----------------------------------

with open("/content/第９回.ipynb","br") as f:
  file = f.read()
print("-----------------ファイル--------------------------")
print(hashlib.md5(file).hexdigest())


#--------------------------１文字の場合--------------------------
#32文字の文字列
#連続する数値だがハッシュ値がまったく違うこと

print("-----------------------１文字の場合--------------------------")
print(hashlib.md5("1".encode("UTF-8")).hexdigest())
```

```python
#Python のオブジェクトを シリアライズ（保存可能なバイト列に変換） するためのライブラリ
#Python のリスト、辞書、クラスなどをファイルに保存 ,後で読み込んで復元

#pickle → データをバイト列にする

import pickle

print("-----連続する数値だがハッシュ値がまったく違うこと--------")
print(hashlib.md5(pickle.dumps(1)).hexdigest())
print(hashlib.md5(pickle.dumps(2)).hexdigest())


def to_byte_hash(val):
  byte_data = pickle.dumps(val)
  return hashlib.md5(byte_data).hexdigest()

print("------------------先頭要素を変更し元に戻す-------------------")
array_10k = list(range(10000))
print(to_byte_hash(array_10k))

#先頭要素を変更
array_10k[0] = -1
print(to_byte_hash(array_10k))

#元に戻す
array_10k[0] = 0
print(to_byte_hash(array_10k))
```

```python
#ビットコインのマイニングは、「ブロックのハッシュ値が一定条件を満たすまで、nonce を総当たりで探す作業」。
#マイニング = 宝くじを引き続けて、当たりが出るまで試す作業

#ビットコインでいう nonce（ナンス)マイニング時にハッシュ値を変えるために何度も書き換えられる特別な数だよ。

import hashlib

def block_to_hash(block_data,nonce):
  #block_data + nonce を文字列にする
  input_str = str(block_data) + str(nonce)

  #SHA-256 を計算するこれはビットコインでいう「ブロックヘッダをハッシュする」
  h=hashlib.sha256(input_str.encode("UTF-8")).hexdigest()

  #先頭から０がいくつかを数える
  cnt = 0
  for v in h:
    if v == "0":
      cnt += 1
    else:
      break
  return h,cnt

#先頭の0が５つの並んだハッシュ値を探す
my_block = "prev_block_tx0_tx1_tx2"
c=1
while True:
  hash_val ,cnt = block_to_hash(my_block,c)
  if cnt == 5:
    print(f"{c}回目の計算で成功しました")
    print(hash_val)
    break
  c +=1
```
